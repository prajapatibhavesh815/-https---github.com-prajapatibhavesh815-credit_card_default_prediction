```python
# histrograms for numerical features
numerical_features = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

plt.figure(figsize=(20,15))
for i,feature in enumerate(numerical_features):
    plt.subplot(5,3,i+1)
    sns.histplot(data[feature],kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout() 
```

```python
# correlation heatmap
plt.figure(figsize=(15,10))
corr_matrix = data.corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation heatmap')
plt.show()
```

```python
# Pair plot for a subset of features to visualize relationships
sampled_data = data.sample(n=500)  # To make the pair plot less dense
pair_plot_features = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'PAY_AMT1', 'PAY_AMT2', 'default.payment.next.month']
sns.pairplot(sampled_data[pair_plot_features], hue='default.payment.next.month', diag_kind='kde', palette='coolwarm')
plt.suptitle('Pair Plot', y=1.02)
```

```python
# independent and dependent features
X=data.drop('LIMIT_BAL',axis=1)
y=data['LIMIT_BAL']
```

```python
X
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
## feature selection based on correlation
X_train.corr()
```

```python
# feature scaling or standardization

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

```python
## Decision Tree Model Training with Hyperparameter tuning
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter ={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
classifier  = DecisionTreeClassifier()
```

```python
clf = GridSearchCV(classifier,param_grid=parameter,cv=5,scoring='accuracy',verbose=3)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
classifier = DecisionTreeClassifier(criterion='gini',max_depth=5,max_features='sqrt',splitter='best')
```

```python
classifier.fit(X_train,y_train)
```

```python
## decision tree prediction
y_pred = classifier.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```

```python
true_positive = conf_mat[0][0]
false_positive = conf_mat[0][1]
false_negative = conf_mat[1][0]
true_nagative = conf_mat[1][1]
```

```python
accuracy = (true_positive + true_nagative) / (true_positive + false_positive+false_negative+true_nagative)
accuracy
```

```python
accuracy = (true_positive + true_nagative) / (true_positive + false_positive+false_negative+true_nagative)
accuracy
```

```python
precision = true_positive/(true_positive+false_positive)
precision
```

```python
Recall = true_positive/(true_positive + false_negative)
Recall
```

```python
F1_score = 2*(Recall*precision)/ (Recall + precision)
F1_score
```
